```python
from IPython.display import display, clear_output
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

## Append to list

```python
data = []

fig = plt.figure(figsize = (13, 5))
ax = fig.add_subplot(111)

for i in range(5000):
    y = np.sin(i / 37) + 0.55*np.sin(i / (37/2))
    data.append(y)
    if i % 30 == 0:
            ax.clear()
            ax.plot(data)
            ax.set_ylim(-2, 2)
            display(fig)
            clear_output(wait=True)
```

## Predefined array

```python
data = np.zeros((5000, 1))
t = np.arange(len(data))/130

fig = plt.figure(figsize = (13, 5))
ax = fig.add_subplot(111)

for i in range(len(data)):
    data[i] = np.sin(i / 37) + 0.55*np.sin(i / (37/2))
    if i % 30 == 0:
            ax.clear()
            ax.plot(t, data)
            ax.set_ylim(-2, 2)
            display(fig)
            clear_output(wait=True)
```

## Like ECG machine

```python
Fs = 130
N = 8*Fs
data = np.zeros((N, 1))
t = np.arange(N)/Fs

fig = plt.figure(figsize = (13, 5))
ax = fig.add_subplot(111)

for i in range(5000):
    n = i % (N)
    data[n] = np.sin(i / 37) + 0.55*np.sin(i / (37/2))
    if i % 30 == 0:
            ax.clear()
            ax.plot(t, data)
            ax.set_ylim(-2, 2)
            display(fig)
            clear_output(wait=True)
```
